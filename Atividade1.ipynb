```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

```

```python

arquivo = files.upload()
df = pd.read_csv('/content/dados.csv',encoding='latin-1')
print(df.head())
```

```python
df.dropna(inplace = True)

print(df.to_string())
```

```python
plt.scatter(df['Idade'],df['Pontos'],color = 'r',alpha=0.5, label='Pontos')

plt.title('Distribuição de pontos por idade')
plt.xlabel('Idade')
plt.ylabel('Pontos')
plt.grid(True)
plt.savefig("Grafico1")
plt.show()
```

```python
media_pontos_por_idade = df.groupby('Idade')['Pontos'].mean()



plt.scatter(df[df['Sexo'] == 'M']['Idade'], df[df['Sexo'] == 'M']['Pontos'], color='b', alpha=0.5, label='Pontos (Masculino)')
plt.scatter(df[df['Sexo'] == 'F']['Idade'], df[df['Sexo'] == 'F']['Pontos'], color='y', alpha=0.5, label='Pontos (Feminino)')

plt.plot(media_pontos_por_idade.index, media_pontos_por_idade.values, color='r', label='Média de Pontos', linewidth=2)

plt.title('Relação entre Idade e Pontos (com Média)')
plt.xlabel('Idade')
plt.ylabel('Pontos')
plt.grid(True)
plt.legend()
plt.savefig("Grafico2")
plt.show()
```
